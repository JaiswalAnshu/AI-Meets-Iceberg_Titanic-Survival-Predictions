#### Model Construction

The above codes calculate the %age of male and female passengers survival rate in (train.csv) on the basis of Gender . It denotes that ~ 75% Of females have survived the incident and only ~ 19% of males have survived .

By we also need to uncover the patterns based on other details of the passengers as this gender based submission is not enough to predict all the patterns . 

So in order to come up with multiple complex patterns we need to analyse the data with multiple dimensions so that we can come up with multiple predictions . 

Since it is quite difficult to consider several columns at once (or, it would take a long time to consider all possible patterns in many different columns simultaneously), we'll use machine learning to automate this for us.

Here we will be building a machine learning model known as random forest model to uncover our patterns. 

Intro on random forest model : It is a  machine learning algorithm that uses an ensemble(group of) of decision trees to make predictions, combining the results of multiple trees to achieve better accuracy and stability than individual trees ths making the model more accurate .

For now we are considering thefollowing features of the dataset for analysis : "Pclass", "Sex", "SibSp", and "Parch".

The model will construct patterns based on these columns from train.csv to make predictions for test.csv dataset.

```python
# Importing the required Library Random Forest Classifier, which is a machine learning algorithm used to make predictions based on past data.
from sklearn.ensemble import RandomForestClassifier

# Getting the target column i.e the column to be predicted 
y= train_data['Survived']

# Selecting the features for model training  
features =["Pclass", "Sex", "SibSp","Parch"]
X = pd.get_dummies(train_data[features])
X_test =pd.get_dummies(test_data[features])

# Training the machine learning model
model= RandomForestClassifier(n_estimators= 100 , max_depth= 5 , random_state=1)
model.fit(X,y)

# Make predictions on test data 
prediction = model.predict(X_test)

# Saving the prediction to csv file 
output = pd.DataFrame({"PassengerId": test_data.PassengerId, "Survived" : prediction })
output.to_csv('submission.csv',index= False)
print("Your submission was successfully saved!")
```
